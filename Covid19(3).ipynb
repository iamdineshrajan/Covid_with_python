```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/DELL/Downloads/covid_19_india.csv",index_col=0)
```

```python
df.dtypes
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df.info()
```

```python
df=df.set_index("Date")
```

```python
df["Year"]=df.index.year
df["Month"]=df.index.month
df["WeekDays"]=df.index.weekday
df.info()
```

```python
monthend=df.groupby([df.index.year,df.index.month]).last()
monthend
```

```python
cases=monthend[["Cured","Deaths","Confirmed"]]
cases=cases.diff()
cases
```

```python
monthwise=pd.concat([monthend,cases.rename(columns={"Cured":"Monthwise_Cured",
                                                    "Deaths":"Monthwise_Deaths",
                                                    "Confirmed":"Monthwise_Confirmed"})],axis=1)
monthwise
```

```python
monthwise=monthwise.fillna({"Monthwise_Cured":0,"Monthwise_Deaths":0,"Monthwise_Confirmed":1})
monthwise
```
