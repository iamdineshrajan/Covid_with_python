```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df.drop(["Time","ConfirmedIndianNational","ConfirmedForeignNational"],axis=1)
```

```python
tamilnadu=df[df["State/UnionTerritory"]=="Tamil Nadu"]
```

```python
tamilnadu.head(20)
```

```python
tamilnadu=tamilnadu.set_index("Date")
```

```python
tamilnadu.tail(3)
```

```python
tamilnadu.info()
```

```python
tamilnadu["Year"]=tamilnadu.index.year
tamilnadu["Month"]=tamilnadu.index.month
tamilnadu["WeekDays"]=tamilnadu.index.weekday
tamilnadu.info()
```

```python
tamilnadu.tail(2)
```

```python
tamilnadu=tamilnadu.drop(["ConfirmedIndianNational","ConfirmedForeignNational"],axis=1)
```

```python
tamilnadu.head(20)
```

```python
monthend_tn=tamilnadu.groupby([tamilnadu.index.year,tamilnadu.index.month]).last()
```

```python
monthend_tn
```

```python
monthend_tn=monthend_tn.rename(index={"Date":"Year/Month"})
```

```python
monthend_tn
```

```python
cases=monthend_tn[["Cured","Deaths","Confirmed"]]
cases=cases.diff()
cases
```

```python
cases=cases.fillna({"Cured":4,"Deaths":1,"Confirmed":74})
```

```python
cases
```

```python
tamilnadu_monthwise=pd.concat([monthend_tn,cases.rename(columns={"Cured":"Monthwise_Cured",
                                                         "Deaths":"Monthwise_Deaths",
                                                        "Confirmed":"Monthwise_Confirmed"})],axis=1)
tamilnadu_monthwise
```
